```python
import pickle
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.inspection import permutation_importance
```

## Random Forest Feature Importance

```python
with open('randomforest.pkl', 'rb') as f:
    rf_model = pickle.load(f)
columns = pd.read_csv('preprocessed.csv').drop(['price'], axis=1).columns
rf_model.feature_importances_
```

```python
plt.figure(figsize=(15,5))
sns.barplot(x=columns, y=rf_model.feature_importances_)
```

## Pemutation Feature Importance

```python
data = pd.read_csv('preprocessed.csv')
y = data['price'].values
y = y.reshape(-1,1)
X = data.drop(['price'], axis=1)
r = permutation_importance(rf_model, X, y, scoring='r2', n_repeats=50)
r
```

```python
plt.figure(figsize=(15,5))
sns.barplot(x=columns, y=r['importances_mean'])
```
